## __Project Requirements (Full Pipeline Flow)__

* First we create a list `X` which contains numbers from **1 to 100**  

* Then we divide `X` into **10 sublists** like:  
  `X1, X2, X3, X4, X5, X6, X7, X8, X9, X10`  
  where each sublist contains a chunk of the main list  


* Then we create a main folder named **"Output"**  
  which will store all files and folders of the project  


* Then we create a file called **master.csv**  
  which stores original data from list `X`  

* This file has **two columns**:  
  * `X` → original numbers  
  * `Square` → square of numbers (to be filled later)  


* Then we create **10 CSV files**  
  each file stores one sublist of `X`  

  Example:  
  * csv_1.csv → X1  
  * csv_2.csv → X2  
  * ...  
  * csv_10.csv → X10  

* Then we create **10 folders inside Output folder**  

  Example:  
  * Output/folder_1  
  * Output/folder_2  
  * ...  
  * Output/folder_10  


* Then we assign CSV files into folders based on a rule  
  (each folder contains 3 files using a structured combination system)  


* Then we fetch data from each CSV file in order:  
  `csv_1.csv → csv_2.csv → ... → csv_10.csv`  

* For each number we:
  * calculate its square (`X²`)  
  * store it back into `master.csv` under the **Square column**  

* If a file is NOT found in one folder:  
  * we search in other folders  
  * until the file is found  

---

 __Final Goal__

✔ Split data into parts  
✔ Distribute across folders  
✔ Recover files if missing  
✔ Compute squares  
✔ Update master dataset  

### __Import Librires__

```python
import os
import csv
import shutil
```

---
* `import os` help us to perform operations on computer memory like `file path`,`join paths` etc.
  
* `import csv` help us to perform operation on csv file like `Create csv file`,`update csv file`,`Read Csv file` and `Write csv file`.

  
* `import shutil` help us to move data from place to other place like `Copy()` , `Paste`,`Move`

---

### __Creating main folder__

```python
main_folder_name = "Output"
os.makedirs(main_folder_name, exist_ok=True)
```

---
* `main_folder_name = "Output"` here we declare the main folder name as __Ouput__.
  
* `os.makedirs(main_folder_name, exist_ok=True)` here we make the folder using folder path and use `exist_ok=True` to avoid duplicate declaration of folder.

---

### __Creating main list and master file path__

```python
x = list(range(1, 101))
master_file_path = os.path.join(main_folder_name, 'master.csv')
```

---
* `x = list(range(1, 101))` here we make a list of numbers from range __1 to 100__.
  
* `master_file_path = os.path.join(main_folder_name, 'master.csv')` here we join the path of main_folder `Ouput` and `master.csv` result `Output/master.csv`

---

### __Creating master.csv file__

```python
with open(master_file_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["number", "square"])
    for num in x:
        writer.writerow([num])
```

---

* `with open(master_file_path, "w", newline="") as f:` here we give file path , set `"W"` to write the csv file and set `newline=''` to avoid overwrite.
  
* ` writer = csv.writer(f)` import __writer from csv__  to perform actions on csv file

  
* `writer.writerow(["number", "square"])` import __writerow__ from writer and set the columns names `number` and `square`.

  
* `for num in x:` run loop until x index value is :-1

  
* `writer.writerow([num])` add data to csv file __rowwise__

---

### __Creating sublists (chunking the main list)__

```python
chunks = 10
sublists = [x[i:i+chunks] for i in range(1, len(x), chunks)]
sublists
```

---
* `chunks = 10` set the value of sublists.
  
* `sublists = [x[i:i+chunks] for i in range(1, len(x), chunks)]` making sublists
---

### __Creating CSV files from sublists__

```python
for i, sublist in enumerate(sublists, start=1):
    filename = f'csv_{i}.csv'
    file_path = os.path.join(main_folder_name, filename)
    with open(file_path, "w", newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['X'])
        for j in sublist:
            writer.writerow([j])
```

---
* `for i, sublist in enumerate(sublists, start=1):` here we loop through all sublists, where `i` is the file number (1, 2, 3...) and `sublist` is each group of data .
  
* `filename = f'csv_{i}.csv'` here we create a file name dynamically like csv_1.csv, csv_2.csv, etc

    
* `file_path = os.path.join(main_folder_name, filename)` here we join folder name and file name to create full file path safely

  
* `with open(file_path, "w", newline='') as f:` here we open the file in write mode, if file exists it will be overwritten

  
* `writer = csv.writer(f)` here we create a CSV writer object to write data into the file

  
* `writer.writerow(['X'])` here we write the header row with column name X

  
* `for j in sublist:` here we loop through each value inside the sublist

  
* `writer.writerow([j])` here we write each value as a new row in the CSV file

---

### __Creating file list and folder assignment list__

```python
files = list(range(1, 11)) 
folder_assignments = []
```

---
* `files = list(range(1, 11))` here we create a list of numbers from 1 to 10 which represents file IDs (csv_1 to csv_10)
  
* `folder_assignments = []` here we create an empty list which will later store combinations of files assigned to each folder

---

### __Creating folder assignments using circular combinations__

```python
for i in range(len(files)):
    combo = [
        files[i % len(files)],
        files[(i + 1) % len(files)],
        files[(i + 2) % len(files)]
    ]
    folder_assignments.append(combo)
folder_assignments
```

---
* `for i in range(len(files)):` here we loop from 0 to 9 (total number of files) to create 10 folder combinations
  
* `combo = [` here we create a list of 3 files for each folder using circular indexing

  
* `files[i % len(files)]` here we select the current file using modulo to avoid index error and loop back to start when reaching end

  
* `files[(i + 1) % len(files)]` here we select the next file in sequence, wrapping around if needed

  
* `files[(i + 2) % len(files)]` here we select the next-next file, also using circular logic

  
* `folder_assignments.append(combo)` here we store this group of 3 files into the folder assignment list

  
* `folder_assignments` here we display final list of all folder combinations

---

### __Creating folders and copying CSV files__

```python
for i, combo in enumerate(folder_assignments, start=1):
    sub_folder_path = os.path.join(main_folder_name, f'folder_{i}')
    os.makedirs(sub_folder_path, exist_ok=True)

    for file_num in combo:
        src = os.path.join(main_folder_name, f'csv_{file_num}.csv')
        dst = os.path.join(sub_folder_path, f'csv_{file_num}.csv')

        if os.path.exists(src):
            shutil.copy(src, dst)
```

---
* `for i, combo in enumerate(folder_assignments, start=1):` here we loop through each folder assignment where `i` is folder number and `combo` is list of files assigned to that folder
  
* `sub_folder_path = os.path.join(main_folder_name, f'folder_{i}')` here we create full path for each folder like Output/folder_1, Output/folder_2, etc

  
* `os.makedirs(sub_folder_path, exist_ok=True)` here we create the folder, and `exist_ok=True` avoids error if folder already exists

  
* `for file_num in combo:` here we loop through each file number inside the current combination

  
* `src = os.path.join(main_folder_name, f'csv_{file_num}.csv')` here we define the source file path (original CSV file location)

  
* `dst = os.path.join(sub_folder_path, f'csv_{file_num}.csv')` here we define destination path where file will be copied inside the folder

  
* `if os.path.exists(src):` here we check if the file exists before copying to avoid errors

  
* `shutil.copy(src, dst)` here we copy the file from source folder to destination folder without deleting original file

---

### __Deleting CSV files after processin__

```python
for i in range(1, 11): 
    src = os.path.join(main_folder_name, f'csv_{i}.csv')
    
    if os.path.exists(src):
        os.remove(src)
```

---
* `for i in range(1, 11):` here we loop through file numbers from 1 to 10
  
* `src = os.path.join(main_folder_name, f'csv_{i}.csv')` here we create the full path of each CSV file inside the main folder

  
* `if os.path.exists(src):` here we check if the file exists before deleting it to avoid errors

  
* `os.remove(src)` here we delete (remove) the file permanently from the folder

### __Finding a file in multiple folders__

```python
def find_file(filename, folders, main_folder):
    for folder in folders:
        path = os.path.join(main_folder, folder, filename)
        if os.path.exists(path):
            return path
    return None
```

---
* `def find_file(filename, folders, main_folder):` here we define a function that searches for a file inside multiple folders
  
* `for folder in folders:` here we loop through each folder one by one to search the file

* `path = os.path.join(main_folder, folder, filename)` here we create the full path of the file inside the current folder

* `if os.path.exists(path):` here we check if the file exists at that location  
* `return path` here we return the file path immediately when the file is found

* `return None` here we return None if the file is not found in any folder

---

###   __Getting folder list and creating result dictionary__

```python
folders = [d for d in os.listdir(main_folder_name) if os.path.isdir(os.path.join(main_folder_name, d))]
number_to_square = {}
```

---
* `folders = [d for d in os.listdir(main_folder_name) if os.path.isdir(os.path.join(main_folder_name, d))]` here we get all items inside the main folder and filter only directories (folders), so we get a list of all subfolders
  
* `number_to_square = {}` here we create an empty dictionary which will store numbers as keys and their squares as values

  ---

### __Reading and updating master CSV file__

```python
for i in range(1, 12):
    filename = f"csv_{i}.csv"
    file_path = find_file(filename, folders, main_folder_name)

    if file_path:
        with open(file_path, "r") as f:
            reader = csv.DictReader(f)
            for row in reader:
                num = int(row["X"])
                number_to_square[num] = num ** 2
```

---
* `for i in range(1, 12):` here we loop through file numbers from 1 to 11 to process each CSV file
  
* `filename = f"csv_{i}.csv"` here we create the file name dynamically like csv_1.csv, csv_2.csv, etc

  
* `file_path = find_file(filename, folders, main_folder_name)` here we search for the file inside all folders using the `find_file` function

  
* `if file_path:` here we check if the file was found before opening it  \

  
* `with open(file_path, "r") as f:` here we open the file in read mode

  
* `reader = csv.DictReader(f)` here we read the CSV file as a dictionary where column names are keys

   
* `for row in reader:` here we loop through each row in the CSV file

  
* `num = int(row["X"])` here we extract the value from column **X** and convert it into an integer

  
* `number_to_square[num] = num ** 2` here we calculate the square of the number and store it in a dictionary

  ---

### __Reading and updating master CSV file__

```python
with open(master_file_path, "r") as f:
    rows = list(csv.DictReader(f))

with open(master_file_path, "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["number", "square"])
    writer.writeheader()
    for row in rows:
        num = int(row["number"])
        row["square"] = number_to_square.get(num, "")
        writer.writerow(row)
```

---
* `with open(master_file_path, "r") as f:` here we open the master CSV file in read mode
  
* `rows = list(csv.DictReader(f))` here we read all rows from the CSV file and convert them into a list of dictionaries


* `with open(master_file_path, "w", newline="") as f:` here we open the same file in write mode which will overwrite the existing content  


* `writer = csv.DictWriter(f, fieldnames=["number", "square"])` here we create a CSV writer and define column names as **number** and **square**

  
* `writer.writeheader()` here we write the header row into the file

  
* `for row in rows:` here we loop through each row from the old data

  
* `num = int(row["number"])` here we extract the number value from each row

  
* `row["square"] = number_to_square.get(num, "")` here we find the square of the number from dictionary and assign it to the row (if not found, keep empty)

  
* `writer.writerow(row)` here we write the updated row back into the CSV file
